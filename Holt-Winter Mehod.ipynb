### Detection of seasonality

```python
moving_avg = data['Close'].rolling(7).mean()
```

```python
plt.figure(figsize=(15,5))
plt.plot(data['Close'])
plt.plot(moving_avg)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(data['Close'])
plt.plot(data['Close']/moving_avg)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(data['Close'])
plt.plot(data['Close']- moving_avg)
plt.show()
```

```python
## Holt-Winter Method
from statsmodels.tsa.api import ExponentialSmoothing
```

```python
#trend is additive and seasonal is multiplicative
fit1 = ExponentialSmoothing(train['Close'], seasonal_periods=12, trend='add', seasonal='mul').fit( use_boxcox=True)
fit2 = ExponentialSmoothing(train['Close'], seasonal_periods=12, trend='add', seasonal='mul', damped=True).fit(use_boxcox=True)
```

```python
fitted1=pd.Series(fit1.fittedvalues, index=train.index)
forecast1=list(fit1.forecast(len(test)))
holt_1=pd.DataFrame(index=test.index,columns=['Prediction'])
holt_1['Prediction']=forecast1
```

```python
plt.figure(figsize=(15,5))
plt.plot(train['Close'], color='Green', label='Train')
plt.plot(test['Close'],color='Blue', label='Test')
plt.plot(fitted1,linestyle='--',color='Red', label="Fitted value")
plt.plot(holt_1['Prediction'], linestyle='--',color='orange', label='forecast')
plt.legend(loc='best')
plt.title("Forecasting price with Holt-Winter method")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rms= np.sqrt(mean_squared_error(test['Close'],holt_1['Prediction']))
print(rms)
```

```python
##using 2nd combination
fitted2=pd.Series(fit2.fittedvalues, index=train.index)
forecast2=list(fit2.forecast(len(test)))
holt_2=pd.DataFrame(index=test.index,columns=['Prediction'])
holt_2['Prediction']=forecast2
```

```python
plt.figure(figsize=(15,5))
plt.plot(train['Close'], color='Green',label ='Train')
plt.plot(test["Close"],color='Blue', label='Test')
plt.plot(fitted2,linestyle='--',color='Red', label='Fitted values')
plt.plot(holt_2['Prediction'], linestyle='--',color='orange', label='Prediction')
plt.legend(loc='best')
plt.title("Forecasting price with Holt-Winter method")
plt.show()
```

```python
rms= np.sqrt(mean_squared_error(test['Close'],holt_2['Prediction']))
print(rms)
```
